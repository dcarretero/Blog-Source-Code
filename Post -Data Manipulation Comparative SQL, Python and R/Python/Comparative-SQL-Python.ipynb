```python
import pandas as pd

airports = pd.read_csv('c:/tmp/data/airports.csv')
airport_frequencies = pd.read_csv('c:/tmp/data/airport-frequencies.csv')
runways = pd.read_csv('c:/tmp/data/runways.csv')
```

**********************************************
        2) SELECT WITH MULTIPLE CONDITIONS
**********************************************

```python
#2.1) Multiple condition, first example

airports[(airports.iso_region == 'US-CA') & (airports.type == 'seaplane_base')]
```

```python
#2.2) Multiple condition, second example

airports[(airports.iso_region == 'US-CA') & (airports.type == 'large_airport')][['ident', 'name', 'municipality']]
```

**********************************************
        3) ORDER BY
**********************************************

```python
#3.1) Order by asc

airport_frequencies[airport_frequencies.airport_ident == 'KLAX'].sort_values('type')
```

```python
#3.2) Order by asc one column and desc the other one

airport_frequencies[airport_frequencies.airport_ident == 'KLAX'].sort_values('type', ascending=False)
```

***********************************************
        4)  IN...NOT IN
***********************************************

```python
#4.1) IN several values
airports[airports.type.isin(['heliport', 'balloonport'])]
```

```python
#4.2) NOT IN several values
airports[~airports.type.isin(['heliport', 'balloonport'])]
```

*****************************************************
        5) GROUP BY
*****************************************************

```python
#5.1) Group by with count by ISO_COUNTRY, TYPE
airports.groupby(['iso_country', 'type']).size()
```

```python
#5.2) GROUP BY ISO_COUNTRY and TYPE with COUNT and ordered ascendent and descendent
#     by ISO_COUNTRY and SIZE
airports.groupby(['iso_country', 'type']).size().to_frame('size').reset_index().sort_values(['iso_country', 'size'], ascending=[True, False])
```

```python
#5.3) Group by more than one columm
airports.groupby(['iso_country', 'type']).size()
```

**********************************************************
        6) LIMIT
**********************************************************

```python
#6.1) Top 10 records
airports.groupby(['iso_country']).size().to_frame('size').nlargest(10, ['size'])
```

```python
# 6.2) Top 10 records with an offset of 10
airports.groupby(['iso_country']).size().to_frame('size').nlargest(20, ['size']).tail(10)
```

******************************************************
        7) HAVING
******************************************************

```python
# 7.1) Types with count over 1000
airports[airports.iso_country == 'US'].groupby('type').filter(lambda g: len(g) > 1000).groupby('type').size().sort_values(ascending=False)
```

**************************************************
        8) MIN,MAX,MEAN AND MEDIAN
**************************************************

```python
# 8.2) Min,max,mean and median
runways.agg({'length_ft': ['min', 'max', 'mean', 'median']})
```

***********************************************************************
        9) JOIN (INNER, LEFT, RIGHT AND OUTER)
***********************************************************************        

```python
# 9.1) INNER JOIN BY AIRPORT REF
airport_frequencies.merge(airports[airports.ident == 'KLAX'][['id']], left_on='airport_ref', right_on='id', how='inner')[['airport_ident', 'type', 'description', 'frequency_mhz']]
```

**********************************************************************
        10) UNION AND UNION ALL
**********************************************************************

```python
# 10.1) UNION ALL
pd.concat([airports[airports.ident.isin(['KLAX','KLGB'])][['name', 'municipality']], airports[airports.ident == 'KLGB'][['name', 'municipality']]])
```

```python
# 10.2) UNION -- adding drop_duplicates() we eliminate duplicate registers taking the same result as UNION
pd.concat([airports[airports.ident.isin(['KLAX','KLGB'])][['name', 'municipality']], airports[airports.ident == 'KLGB'][['name', 'municipality']]]).drop_duplicates()
```

**********************************************************************
        11) EXCEPT
**********************************************************************

```python
# 11.1) Except between two dataframes
airports_df1=airports[airports.ident.isin(['KLAX', 'KLGB'])]
airports_df2=airports[airports.ident.isin(['KLGB'])]
airports_df1[~airports_df1.id.isin(airports_df2.id)]['name','municipality','ident']
```
